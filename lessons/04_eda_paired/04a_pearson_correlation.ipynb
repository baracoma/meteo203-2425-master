```python
import numpy as np
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt

# Generating synthetic data with different correlations
np.random.seed(42)
n = 100

# Perfect positive correlation
x_perfect_pos = np.linspace(0, 100, n)
y_perfect_pos = x_perfect_pos

# Strong positive correlation
x_strong_pos = np.linspace(0, 100, n)
y_strong_pos = x_strong_pos * 0.8 + np.random.normal(0, 10, n)

# Weak positive correlation
x_weak_pos = np.linspace(0, 100, n)
y_weak_pos = x_weak_pos * 0.3 + np.random.normal(0, 30, n)

# No correlation
x_none = np.linspace(0, 100, n)
y_none = np.random.normal(50, 25, n)

# Weak negative correlation
x_weak_neg = np.linspace(0, 100, n)
y_weak_neg = -0.3 * x_weak_neg + np.random.normal(0, 30, n)

# Strong negative correlation
x_strong_neg = np.linspace(0, 100, n)
y_strong_neg = -0.8 * x_strong_neg + np.random.normal(0, 10, n)

# Create a DataFrame for easier plotting and analysis
df = pd.DataFrame({
    'x_perfect_pos': x_perfect_pos, 'y_perfect_pos': y_perfect_pos,
    'x_strong_pos': x_strong_pos, 'y_strong_pos': y_strong_pos,
    'x_weak_pos': x_weak_pos, 'y_weak_pos': y_weak_pos,
    'x_none': x_none, 'y_none': y_none,
    'x_weak_neg': x_weak_neg, 'y_weak_neg': y_weak_neg,
    'x_strong_neg': x_strong_neg, 'y_strong_neg': y_strong_neg
})

# List of data and titles for the subplots
data_pairs = [
    (df['x_perfect_pos'], df['y_perfect_pos'], "Perfect Positive Correlation"),
    (df['x_strong_pos'], df['y_strong_pos'], "Strong Positive Correlation"),
    (df['x_weak_pos'], df['y_weak_pos'], "Weak Positive Correlation"),
    (df['x_none'], df['y_none'], "No Correlation"),
    (df['x_weak_neg'], df['y_weak_neg'], "Weak Negative Correlation"),
    (df['x_strong_neg'], df['y_strong_neg'], "Strong Negative Correlation")
]

# Set up the 2x3 grid for plotting
fig, axes = plt.subplots(2, 3, figsize=(15, 10), constrained_layout=True)

# Loop through each subplot and plot the data
for i, (x, y, title) in enumerate(data_pairs):
    ax = axes[i // 3, i % 3]
    sns.scatterplot(x=x, y=y, ax=ax)
    correlation = np.corrcoef(x, y)[0, 1]
    ax.set_title(f"{title}\nCorrelation: {correlation:.2f}")
    ax.set_xlabel("X")
    ax.set_ylabel("Y")

plt.suptitle("Scatter Plots with Different Correlation Strengths", fontsize=16)
plt.show()
```

```python
# Set a seed for reproducibility
np.random.seed(42)
n = 50  # Set consistent length for all datasets

# Dataset 1: Linear positive correlation
x1 = np.linspace(0, 10, n)
y1 = 0.5 * x1 + np.random.normal(0, 0.5, size=n)

# Dataset 2: Nonlinear relationship (parabolic)
x2 = np.linspace(0, 10, n)
y2 = (x2 - 5) ** 2 / 10 + np.random.normal(0, 0.2, size=n)

# Dataset 3: Noisy relationship (wide scatter)
x3 = np.linspace(0, 10, n)
y3 = 0.5 * x3 + np.random.normal(0, 2, size=n)

# Dataset 4: Linear relationship with outliers
x4 = np.linspace(0, 10, n)
y4 = 0.5 * x4 + np.random.normal(0, 0.5, size=n)
y4[-3:] = [15, 16, 17]  # Add outliers to the last three points

# Put data in a DataFrame
data = pd.DataFrame({
    'x1': x1, 'y1': y1,
    'x2': x2, 'y2': y2,
    'x3': x3, 'y3': y3,
    'x4': x4, 'y4': y4
})

# Set up a 2x2 grid for plotting
fig, axes = plt.subplots(2, 2, figsize=(12, 10), constrained_layout=True)

# List of data pairs and titles for subplots
data_pairs = [
    (data['x1'], data['y1'], "Dataset 1: Linear Positive Correlation"),
    (data['x2'], data['y2'], "Dataset 2: Nonlinear (Parabolic) Relationship"),
    (data['x3'], data['y3'], "Dataset 3: Noisy Linear Relationship"),
    (data['x4'], data['y4'], "Dataset 4: Linear with Outliers")
]

# Loop through each subplot
for i, (x, y, title) in enumerate(data_pairs):
    ax = axes[i // 2, i % 2]
    sns.scatterplot(x=x, y=y, ax=ax)
    correlation = np.corrcoef(x, y)[0, 1]
    ax.set_title(f"{title}\nPearson Correlation: {correlation:.2f}")
    ax.set_xlabel("X")
    ax.set_ylabel("Y")

plt.suptitle("Examples of Dangers of Pearson Correlation", fontsize=16)
plt.show()
```
